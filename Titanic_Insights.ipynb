Concepts Studied
1. pd.read_csv()

Used to load a CSV file into a Pandas DataFrame.

2. df.info()

Displays the dataset's structure, including column names, data types, and missing values.

3. df.describe()

Provides statistical summaries such as mean, median, minimum, maximum, and standard deviation for numerical columns.

4. df.isnull()

Identifies missing values in the dataset.

5. df.dropna()

Removes rows or columns containing missing values.

6. Filtering & Slicing

  Used to select specific rows or columns from the dataset based on conditions or positions.

7. groupby()

Groups data based on one or more columns to perform analysis on each group.

8. agg()

Applies one or more aggregate functions such as mean, sum, count, or maximum on grouped data.

9. value_counts()

Counts the frequency of unique values in a column.

10. df.corr()

Calculates the correlation between numerical columns to identify relationships.

The dropna() function removes rows containing missing values. In the Titanic dataset, many rows contain missing values in the Cabin column, so using dropna() removes a large portion of the data. Therefore, while dropna() is useful to understand, it is not always the best choice for this dataset.

```python
# Passengers older than 30
df[df["Age"] > 30]
```

```python
# Only female passengers
df[df["Sex"] == "female"]
```

```python
# Passengers who survived
df[df["Survived"] == 1]
```

```python
# Female passengers who survived
df[(df["Sex"] == "female") & (df["Survived"] == 1)]
```

Observation:
 Filtering the dataset allows us to extract specific groups of passengers based on conditions such as age, gender, and survival status. This helps in analyzing particular categories and identifying meaningful patterns within the data

```python
# Select first 5 rows
df.iloc[0:5]
```

```python
df.groupby("Sex")["Age"].mean()
```

```python
df.groupby("Pclass").agg({
    "Age": "mean",
    "Fare": "mean"
})
```

```python
df["Survived"].value_counts()
```

```python
df.corr(numeric_only=True)
```
